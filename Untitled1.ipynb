#### importing libraries

```python
import pandas as pd
from bs4 import BeautifulSoup as bs4
import matplotlib.pyplot as plt
import numpy as np
import requests
import csv
```

```python
num_of_pages = 10
pages = []
```

#### getting data

```python
for i in range(1, num_of_pages+1):
    res = requests.get(f'http://books.toscrape.com/catalogue/page-{i}.html')
    soup = bs4(res.content, 'lxml')
    page = soup.find_all('article', class_='product_pod')
    pages.append(page)
```

#### parsing data & create csv file

```python
with open('books.csv', 'w', newline='') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(['book_title','book_url','img_url','book_rating','book_price'])

    for page in pages:
        for article in page:
            book_title = article.select('h3 a')[0].get('title')
            book_url = article.select('.image_container a')[0].get('href')
            img_url = article.select('.image_container a img')[0].get('src')
            book_rating = article.find('p', class_='star-rating').get('class')[1]
            book_price = article.select('.product_price p.price_color')[0].get_text()
            
            csv_writer.writerow([book_title, book_url, img_url, book_rating, book_price])
```

#### read csv file

```python
book_df = pd.read_csv('books.csv', encoding='cp1252')
```

#### plotting number of books by star rating

```python
rating = book_df['book_rating'].value_counts()
```

```python
objects = []
performance = []

for index,value in rating.items():
    objects.append(index)
    performance.append(value)
    
y_pos = np.arange(len(objects))

plt.bar(y_pos, performance, align='center')
plt.xticks(y_pos, objects)
plt.ylabel('no of books')
plt.xlabel('stars')
plt.title('Star rating of books')

plt.show()
```

#### sort dataframe by book price

```python
sorted_df = book_df.sort_values(by=['price'], ascending=False)
```

```python
sorted_df.head(10)
```

#### plot top 10 books by price

```python
book_name = [book.split(':')[0].split('(')[0] for book in sorted_df.head(10)['book_title']]
b_price = [float(price) for price in sorted_df.head(10)['price']]
```

```python
book_name.reverse()
```

```python
b_price.reverse()
```

```python
plt.style.use('ggplot')
plt.barh(book_name, b_price, color='#f43ad2')
plt.yticks(book_name)
plt.xlabel('book price')
plt.title('top 10 expensive books')
for index, value in enumerate(b_price):
    plt.text(value + 1, index, str(value), color='blue', fontweight='bold')
plt.show()
```
